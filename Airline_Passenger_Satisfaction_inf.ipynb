## Libraries

```python
import pandas as pd
import numpy as np
import pickle 
import json
```

## Model Load

```python
with open("list_num_columns.txt", 'r') as file_1:
    list_num_skew_columns = json.load(file_1)

with open("list_cat_nominal_columns.txt", "r") as file_2:
    nom_col_skew = json.load(file_2)

with open("list_cat_ordinal_columns.txt", "r") as file_3:
    ord_col_skew = json.load(file_3)

with open("best_pipeline.pkl", "rb") as file_4:
    best_pipeline = pickle.load(file_4)
```

## Data Inference

```python
data_inf={'id': {0: 701232, 1: 50007},
 'gender': {0: 'Female', 1: 'Male'},
 'customer type': {0: 'Loyal Customer', 1: 'disloyal Customer'},
 'age': {0: 19, 1: 25},
 'type of travel': {0: 'Personal Travel', 1: 'Business travel'},
 'class': {0: 'Business', 1: 'Eco Plus'},
 'flight distance': {0: 460, 1: 235},
 'inflight wifi service': {0: 2, 1: 3},
 'departure/arrival time convenient': {0: 4, 1: 2},
 'ease of online booking': {0: 3, 1: 3},
 'gate location': {0: 1, 1: 3},
 'food and drink': {0: 1, 1: 1},
 'online boarding': {0: 3, 1: 3},
 'seat comfort': {0: 4, 1: 1},
 'inflight entertainment': {0: 5, 1: 1},
 'on-board service': {0: 4, 1: 1},
 'leg room service': {0: 3, 1: 5},
 'baggage handling': {0: 3, 1: 3},
 'checkin service': {0: 4, 1: 1},
 'inflight service': {0: 5, 1: 4},
 'cleanliness': {0: 5, 1: 1},
 'departure delay in minutes': {0: 15, 1: 1},
 'arrival delay in minutes': {0: 10.0, 1: 6.0}
}
```

```python
data_inf = pd.DataFrame(data_inf)

data_inf.info()
data_inf
```

```python
#Data creation
age_category = []
for x in data_inf["age"]:
    if 6 <= x <= 21:
        age_category.append('Generation z') 
    elif 22 <= x <= 36:
        age_category.append('Millennials')
    elif 37 <= x <= 52:
        age_category.append('Generation X')
    elif 53 <= x <= 73:
        age_category.append('Baby Boomers')
    else:
        age_category.append('Silent Generation')

data_inf["generation"] = age_category
```

```python
# Split between numeric and categoric column
data_inf_num_skew = data_inf[list_num_skew_columns]
data_inf_cat_nom = data_inf[nom_col_skew]
data_inf_cat_ord = data_inf[ord_col_skew]

display(data_inf_num_skew)
display(data_inf_cat_nom)
display(data_inf_cat_ord)
```

```python
#predict satisfied using pipeline
y_predict_inf = best_pipeline.predict(data_inf)
```

```python
# add result to dataframe
data_inf["satisfaction"] = y_predict_inf
```
